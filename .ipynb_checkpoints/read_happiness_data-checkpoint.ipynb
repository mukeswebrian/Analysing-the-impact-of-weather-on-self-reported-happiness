```python
import pandas as pd
```

```python
happiness = pd.read_excel("geographicbreakdownreferencetable_tcm77-417203.xls",
                         sheet_name="Happiness", header=5)
```

```python
happiness.head()
```

```python
to_keep = ["Area Codes", "Area names", "Unnamed: 2", "Unnamed: 3",
          "Unnamed: 4", "Unnamed: 5", "Unnamed: 6","Unnamed: 7",
          "Unnamed: 8","Unnamed: 24"]

# Keep only columns of interest
for column in happiness.columns:
    if not column in to_keep:
        happiness.drop(columns=[column], inplace=True)
```

```python
rename = {"Area Codes" : "area_codes",
          "Area names" : "area_names",
          "Unnamed: 2" : "county",
          "Unnamed: 3" : "sub_county",
          "Unnamed: 4" : "low_0_4",
          "Unnamed: 5" : "medium_5_6",
          "Unnamed: 6" : "high_7_8",
          "Unnamed: 7" : "vhigh_9_10",
          "Unnamed: 8" : "avg_rating",
          "Unnamed: 24": "sample_size"}

# Rename data columns
happiness.columns = rename.values()
happiness.head(10)
```

```python
happiness.tail(20)
```

```python
# Drop trailing rows
for i in range(441, len(happiness)):
    happiness.drop(index=[i], inplace=True)
    
happiness.tail(10)
```

```python
happiness.describe()
```

```python
# Get Area codes of interest
columns = ["area_code", "area_name", "latitude", "longitude"]
regions = pd.read_csv("regions.txt", names=columns)
regions
```

```python
for i in range(0, len(happiness)):
    
    if not happiness.area_codes.loc[i] in regions.area_code.values:
        happiness.drop(index=[i], inplace=True)
```

```python
happiness.reset_index(inplace=True)
happiness
```

```python
happiness.drop(columns=["index"], inplace=True)
happiness
```

```python
happiness.drop(columns=["county", "sub_county"], inplace=True)
happiness
```
