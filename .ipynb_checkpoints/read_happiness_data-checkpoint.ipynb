```python
import pandas as pd
```

```python
happiness = pd.read_excel("geographicbreakdownreferencetable_tcm77-417203.xls",
                         sheet_name="Happiness", header=5)
```

```python
happiness.head()
```

```python
to_keep = ["Area Codes", "Area names", "Unnamed: 2", "Unnamed: 3",
          "Unnamed: 4", "Unnamed: 5", "Unnamed: 6","Unnamed: 7",
          "Unnamed: 8","Unnamed: 24"]

# Keep only columns of interest
for column in happiness.columns:
    if not column in to_keep:
        happiness.drop(columns=[column], inplace=True)
```

```python
rename = {"Area Codes" : "area_codes",
          "Area names" : "area_names",
          "Unnamed: 2" : "county",
          "Unnamed: 3" : "sub_county",
          "Unnamed: 4" : "low_0_4",
          "Unnamed: 5" : "medium_5_6",
          "Unnamed: 6" : "high_7_8",
          "Unnamed: 7" : "vhigh_9_10",
          "Unnamed: 8" : "avg_rating",
          "Unnamed: 24": "sample_size"}

# Rename data columns
happiness.columns = rename.values()
happiness.head(10)
```

```python
happiness.tail(20)
```

```python
# Drop trailing rows
for i in range(441, len(happiness)):
    happiness.drop(index=[i], inplace=True)
    
happiness.tail(10)
```

```python
happiness.describe()
```

```python
# Get Area codes of interest
columns = ["area_code", "area_name", "latitude", "longitude"]
regions = pd.read_csv("regions.txt", names=columns)
regions
```

```python
for i in range(0, len(happiness)):
    
    if not happiness.area_codes.loc[i] in regions.area_code.values:
        happiness.drop(index=[i], inplace=True)
```

```python
happiness.reset_index(inplace=True)
happiness
```

```python
happiness.drop(columns=["index"], inplace=True)
happiness
```

```python
happiness.drop(columns=["county", "sub_county"], inplace=True)
happiness
```

```python
# Merge coordinates with happiness data
happiness = pd.concat([happiness, regions], axis=1)
happiness.drop(columns=["area_name", "area_code"], inplace=True)
happiness
```

```python
# Visulize happiness data
import seaborn as sns

# check for coorelation between lattitude and average happiness rating
sns.scatterplot(data=happiness, x="latitude", y="avg_rating")
```

```python
# Store happiness data in a mogodb database
from pymongo import MongoClient

def make_lineObj(line, data):
    
    # create mongo document object
    return {
        "area_codes" :  data.area_codes.iloc[line],
        "area_names" : data.area_names.iloc[line],
        "low_0_4" : data.low_0_4.iloc[line],
        "medium_5_6" : data.medium_5_6.iloc[line],
        "high_7_8" : data.high_7_8.iloc[line],
        "vhigh_9_10" : data.vhigh_9_10.iloc[line],
        "avg_rating" : data.avg_rating.iloc[line],
        "sample_size" : data.sample_size.iloc[line]
    }

def storeObj(obj):
    collection = "happiness_data"
    
    # Config database server
    client = MongoClient(host="localhost", port=27017)
    db = client.pitds_weather_data
    weather_data = db[collection]
    
    # Store document
    weather_data.insert_one(obj)
    print("successfully stored: ", obj["area_names"])
    
    
for row in range(0, len(happiness)):
    storeObj(make_lineObj(row, happiness))
```
