## analyze happiness data

```python
# Visualize happiness data
%matplotlib inline
%pylab

import matplotlib.pyplot as plt
import seaborn as sns

levels = ["low_0_4", "medium_5_6", "high_7_8", "vhigh_9_10"]
index = 0
ax_index = 0


fig, ax = plt.subplots(6, 1)

for index in range(6, len(happiness)):
    vals =[]
    for level in levels:
        vals.append(happiness[level].iloc[index])

    sns.barplot(x=levels, y=vals, ax=ax[ax_index]).set_ylabel(happiness.area_names.iloc[index])
    ax_index += 1
    
pylab.rcParams['figure.figsize'] = (8, 20)
```

- Based on the skewed results, it is evident that people tend mostly to rank their own happiness levels as either high or very high. 

```python
# Merge coordinates with happiness data
happiness = pd.concat([happiness, regions], axis=1)
happiness.drop(columns=["area_name", "area_code"], inplace=True)
happiness
```

```python

# check for coorelation between lattitude and happiness rating

def plot_scatter(happiness_rating):
    
    rows = 2
    cols = 2
    fig, ax = plt.subplots(rows,cols)

    to_plot = ["sample_size", "latitude", "longitude"]
    plot_index = 0

    for row in range(0, rows):
        for col in range(0, cols):
        
            axes = ax[row][col]
            sns.scatterplot(data=happiness, x=to_plot[plot_index], y=happiness_rating, ax=axes)
            axes.set_xlabel("")
            axes.set_xticklabels("")
            axes.set_yticklabels("")
            axes.set_title(to_plot[plot_index])
            plot_index += 1
        
            if plot_index >= len(to_plot)-1:
                break
```

```python
plot_scatter(happiness_rating="low_0_4")
```

```python
plot_scatter(happiness_rating="medium_5_6")
```

```python
plot_scatter(happiness_rating="high_7_8")
```

```python
plot_scatter(happiness_rating="vhigh_9_10")
```

```python
plot_scatter(happiness_rating="avg_rating")
```

```python
# Store happiness data in a mogodb database
from pymongo import MongoClient

def make_lineObj(line, data):
    ''' create a dict object of a specified row
        of a dataframe
    '''
   
    obj = {}
    
    for column in data.columns:
        obj[column] = data[column].iloc[line]
        
    return obj

def storeObj(obj):
    collection = "happiness_data"
    
    # Config database server
    client = MongoClient(host="localhost", port=27017)
    db = client.pitds_weather_data
    weather_data = db[collection]
    
    # Store document
    weather_data.insert_one(obj)
    print("successfully stored: ", obj["area_names"])
    
    
for row in range(0, len(happiness)):
    storeObj(make_lineObj(row, happiness))
```
